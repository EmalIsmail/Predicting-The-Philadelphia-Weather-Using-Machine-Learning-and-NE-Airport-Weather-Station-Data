Creating a function to make predication

```python
def create_prediction(predictors, core_weather, reg):
    #We will create the train and test set based on the historical data
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]
    
    #we will fit the Regression model
    reg.fit(train[predictors], train["target"])
    
    #Generate Predicaiton
    predications = reg.predict(test[predictors])
    
    #Generate error
    error = mean_absolute_error(test["target"], predications)
    
    #Combining the dataframe
    combined = pd.concat([test["target"], pd.Series(predications, index=test.index)], axis=1)
    combined.columns = ["actual", "predications"]
    return error, combined
    
    
```

Adding in Rolling means
